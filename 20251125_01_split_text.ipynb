#### emojiçš„unicodeä¸€ç›´åœ¨æ›´æ–°,æ‰€ä»¥ç›´æ¥åªè¯†åˆ«æ±‰è¯­éƒ¨åˆ†

```python
def match_chinese(text):
    
    # åŒ¹é…æ‰€æœ‰æ±‰å­—å­—ç¬¦
    chinese_pattern = r'[\u4e00-\u9fa5]+' # \u4e00-\u9fa5 ç”¨äºè¡¨ç¤ºUnicodeä¸­å¸¸è§æ±‰å­—çš„èŒƒå›´
    # æŸ¥æ‰¾æ‰€æœ‰æ±‰å­—
    chinese_text = re.findall(chinese_pattern,text)
    # re.findall(pattern, string): è¿™æ˜¯ Python re (æ­£åˆ™è¡¨è¾¾å¼) æ¨¡å—ä¸­çš„ä¸€ä¸ªæ–¹æ³•ã€‚
    # å®ƒä¼šåœ¨ text å­—ç¬¦ä¸²ä¸­æ‰¾åˆ°æ‰€æœ‰ä¸ chinese_pattern åŒ¹é…çš„éé‡å å­ä¸²ï¼Œå¹¶å°†å®ƒä»¬ä½œä¸ºä¸€ä¸ªåˆ—è¡¨è¿”å›ã€‚
        
    print("åŒ¹é…åˆ°çš„ä¸­æ–‡å­—ç¬¦ï¼š", chinese_text)
    return chinese_text
```

```python
sentence = "æˆ‘å2è·¯è½¦ç¬¬33æ¬¡æ¥åˆ°äº†ç¾ä¸½çš„åŒ—äº¬é¢å’Œå›­goodï¼Œâœ…ğŸ‘Œ"
sentence_match = match_chinese(sentence)
sentence_match 
```

```python
sentence = 'æˆ‘å2è·¯è½¦ç¬¬33æ¬¡æ¥åˆ°äº†ç¾ä¸½çš„åŒ—äº¬é¢å’Œå›­ï¼Œä»¤äººå°è±¡æ·±åˆ»ã€‚....âœ…ğŸ‘Œ'

# å»é™¤æ ‡ç‚¹ç¬¦å·ç­‰
sentence_no_punctuation = remove_punctuation(sentence)

# å»æ‰æ•°å­—
sentence_no_digits = remove_digits(sentence_no_punctuation)

# åŒ¹é…ä¸­æ–‡å­—ç¬¦,å°†åˆ—è¡¨è½¬åŒ–ä¸ºå­—ç¬¦ä¸²
def match_chinese(text):
    # åŒ¹é…æ‰€æœ‰è¿ç»­çš„æ±‰å­—ç‰‡æ®µ
    chinese_pattern = r'[\u4e00-\u9fa5]+'
    chinese_list = re.findall(chinese_pattern, text)
    # å…³é”®ä¿®æ­£ï¼šä½¿ç”¨ join() å°†åˆ—è¡¨åˆå¹¶ä¸ºä¸€ä¸ªå­—ç¬¦ä¸²
    return "".join(chinese_list)
sentence_match = match_chinese(sentence_no_digits)

# ä½¿ç”¨jiebaåˆ†è¯
word_list = jieba.lcut(sentence_match)

# å»é™¤åœç”¨è¯
filtered_words = [word for word in word_list if word not in stop_words]
filtered_words
```
