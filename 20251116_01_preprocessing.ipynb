```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

def GetCookie(url):
    
    driver = webdriver.Chrome()
    driver.maximize_window()  # æœ€å¤§åŒ–çª—å£
    
    # æ‰“å¼€ç½‘é¡µ
    driver.get(url)
    
    time.sleep(50)
    
    # è¾“å…¥ä½ çš„ç”¨æˆ·åå’Œå¯†ç 
    
    #è·å–æ›´æ–°çš„cookie
    driver.refresh()
    
    cookies = driver.get_cookies()

    # è¾“å‡º cookies
    #for cookie in cookies:
    #    print(cookie)
    # å°† cookies è½¬æ¢ä¸º header æ ¼å¼ :  Apache=349490162003.9359.1738739237863
    cookie_string = '; '.join([f"{cookie['name']}={cookie['value']}" for cookie in cookies])
    print(cookie_string)
    return cookie_string
    
url = "https://weibo.com"
cookie_string = GetCookie(url)
```

```python
import os
from selenium.webdriver.chrome.options import Options
import xlrd
from xlutils.copy import copy
import xlwt
#from main import weibo_cols,cols,path

path = 'keyword.xlsx'
weibo_cols = ['å…³é”®è¯','å‘åšç”¨æˆ·','å‘åšæ—¶é—´','åšæ–‡å†…å®¹']

def WriteNewDataToSheet(path,value,data_cols):
        """
        æ’å…¥æ•°æ®åˆ°è¡¨æ ¼é‡Œé¢
        :param path : å…·ä½“çš„æ–‡ä»¶åœ°å€
        :param value : å…·ä½“çš„æ•°æ®
        :param data_cols : æƒ³è¦æ’å…¥çš„æ•°æ®åˆ—
        """
        
        if not os.path.exists(path):  
            # å¦‚æœæ–‡ä»¶ä¸å­˜åœ¨ï¼Œåˆ™æ–°å»º
            new_workbook = xlwt.Workbook()
            new_worksheet = new_workbook.add_sheet("Sheet1")
            for col, col_name in enumerate(data_cols):
                new_worksheet.write(0, col, col_name)  # å†™å…¥åˆ—å
            new_workbook.save(path)
    
        # æ‰“å¼€å·²æœ‰çš„æ–‡ä»¶
        workbook = xlrd.open_workbook(path, formatting_info=True)
        sheets = workbook.sheet_names()
        
        worksheet = workbook.sheet_by_name(sheets[0])  # è·å–ç¬¬1ä¸ª Sheet
        rows_old = worksheet.nrows  # è·å–å·²å­˜åœ¨çš„æ•°æ®è¡Œæ•°

        # å¤åˆ¶ workbook ä»¥è¿›è¡Œå†™å…¥
        new_workbook = copy(workbook)
        new_worksheet = new_workbook.get_sheet(0)

        # æ£€æŸ¥æ˜¯å¦å·²æœ‰ç›¸åŒçš„æ•°æ®
        # ç”¨æˆ·å&æ—¶é—´ç›¸åŒçš„è¯ -ã€‹ åŒä¸€æ¡æ•°æ®
        for i in range(1, rows_old):
            if value[1] == worksheet.cell_value(i, 1) and value[2] == worksheet.cell_value(i, 2): 
                return  # æ•°æ®å·²å­˜åœ¨ï¼Œç›´æ¥è¿”å›

        # æ·»åŠ æ–°æ•°æ®
        for j in range(len(data_cols)):
            new_worksheet.write(rows_old, j, value[j])  # å†™å…¥æ•°æ®

        new_workbook.save(path)  # ä¿å­˜æ–‡ä»¶
    #print("å†™å…¥æˆåŠŸ")
```
