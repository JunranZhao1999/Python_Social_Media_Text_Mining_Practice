### snownlp
SnowNLP 是一个专门针对 中文文本 处理的 Python 库，提供了自然语言处理（NLP）相关的功能，类似于英文的 TextBlob。它支持 分词、词性标注、情感分析、文本摘要、关键词提取、拼音转换 等多种文本处理功能，使用简单，非常适合中文 NLP 任务。

特别的SnowNLP 自带了一个基于豆瓣影评数据训练的情感分析模型。

但是其能力比较初级

s = SnowNLP(text_string)

s.sentiments

s.summary(n)

#### 分析情感

**s1 = SnowNLP(text1)**

snownlp分析的是字符串

**s1.sentiments** 调用 s1 对象的情感分析方法。

**s2.sentiments** 调用 s2 对象的情感分析方法。

SnowNLP 的 sentiments 属性返回一个浮点数，表示文本是**积极（正面）**的可能性。这个值范围在 $0$ 到 $1$ 之间。值越接近 $1$，表示情感越积极（正面）。值越接近 $0$，表示情感越消极（负面）。

- 结果接近 1 表示正向情感
- 结果接近 0 表示负向情感

```python
from snownlp import SnowNLP

text1 = '这个产品真的太棒了，我非常喜欢！'
text2 = '这次体验太糟糕了，我再也不会购买了。'

s1 = SnowNLP(text1)
s2 = SnowNLP(text2)

print(s1.sentiments)
print(s2.sentiments)
```

#### 提取摘要

**s.summary(1)**

s.summary() 是 SnowNLP 对象的一个方法，用于对文本进行自动提取式摘要（Extractive Summarization）。

提取式摘要的工作原理是：它会分析文本中每个句子的重要性（通常使用 TextRank 等算法），并从中选出最能代表原文核心内容的句子作为摘要。

参数 (1) 的含义： 这个数字参数指定了您希望摘要中包含的句子数量。

s.summary(1) 表示从原文中提取并返回 1 句作为摘要。

如果参数是 3，它将返回 3 句最重要的句子。

```python
# 定义了一个多行字符串变量
text = '''
自然语言处理（NLP）是人工智能（AI）的一个重要方向，旨在让计算机理解和生成自然语言。
目前，NLP 主要应用于机器翻译、文本摘要、情感分析等领域。
其中，情感分析可以用于市场分析、舆情监测等方面，具有广泛的应用价值。
'''

s = SnowNLP(text) # snownlp分析的是字符串
s.summary(1) # 提取一句话摘要
```

### Transformers
transformers 是 Hugging Face 开发的 自然语言处理（NLP）库，它提供了 预训练模型 以及 深度学习 API，可以轻松用于：
- 文本分类（如情感分析）
- 命名实体识别（NER）
- 机器翻译
- 文本摘要
- 文本生成（如 ChatGPT）
- 语音和视觉任务（如 CLIP）

主要特点
- 支持多种预训练模型：BERT, GPT, RoBERTa, T5, LLaMA 等。

- Model（模型： 加载 预训练 或 自定义微调 的模型。
- Trainer（训练工具：用于 微调（fine-tuning） 预训练模型。

#### 加载预训练的 BERT 中文分词器（Tokenizer），并对一段中文文本进行分词和编码，将其转换为 PyTorch 张量格式

**from transformers import AutoTokenizer**

在 Hugging Face 库中，AutoTokenizer 是一个非常方便的类，它可以根据您指定的模型名称（如 bert-base-chinese）自动识别并加载该模型对应的正确的分词器。这省去了手动查找和导入特定分词器类的麻烦。

**tokenizer = AutoTokenizer.from_pretrained("bert-base-chinese")**
from_pretrained(...): 这是加载预训练模型的标准方法。

"bert-base-chinese": 这是 Hugging Face 模型仓库中的一个标识符，代表一个基于 BERT 架构、使用简体中文语料库训练的基础模型。

结果： 代码会下载并加载与该模型配套的词汇表文件 (vocab.txt) 和分词规则，创建一个名为 tokenizer 的分词器对象。

**tokens = tokenizer("你好，我是micky moush", return_tensors="pt")**
解释： 这一行将原始文本转换为模型可以理解的数字格式（编码）。

输入文本： "你好，我是micky moush"。

    分词（Tokenization）： 分词器首先将这段文本拆分成一个个“词块”（tokens）。BERT 分词器会对中文进行字级别分词，对英文进行 WordPiece 分词。

    编码（Encoding）： 将每个“词块”映射到其在词汇表中的对应 ID（整数）。同时，它还会自动添加 BERT 模型要求的特殊 token，例如：

        [CLS] (Classification)：放在序列的开头。
    
        [SEP] (Separation)：放在句子末尾。

return_tensors="pt": 这是一个关键参数，它指示分词器将最终的编码结果封装成 PyTorch (pt) 框架使用的张量（Tensor）对象。

```python
from transformers import AutoTokenizer
# 加载预训练的分词器
tokenizer = AutoTokenizer.from_pretrained('bert-base-chinese') 
# 对文本进行分词和编码
tokens = tokenizer('你好，我是micky moush',return_tensors = 'pt') # 封装成tensor张量
tokens
```

#### 中文情感分析

使用 Hugging Face 的 transformers 库和特定的中文预训练模型来对多条中文评论进行批量情感分析

中文情感分析

classifier = pipeline("sentiment-analysis", model="uer/roberta-base-finetuned-jd-binary-chinese")
- "sentiment-analysis"：指定 任务类型 为情感分析。
- model="uer/roberta-base-finetuned-jd-binary-chinese"：加载 UER 预训练的 RoBERTa 模型，该模型在京东商品评论数据上进行了微调，适用于 中文二分类情感分析（正/负）。

**for text, result in zip(texts, results):
    print(f"文本: {text}")
    print(f"情感: {result['label']}, 置信度: {result['score']:.4f}")**

使用 zip 函数将原始文本 (texts) 和对应的分析结果 (results) 配对，并循环打印。

result['label']: 提取模型预测的情感标签（在这个模型中通常是 POSITIVE 或 NEGATIVE）。

result['score']: 提取模型对该标签的置信度。

:.4f: 格式化输出，将置信度四舍五入到小数点后四位，使其更易读。

**置信度是模型对它自己预测结果的把握程度或确定性程度**

```python
from transformers import pipeline

# 加载 BERT 中文情感分析模型
classifier = pipeline('sentiment-analysis',model = 'uer/roberta-base-finetuned-jd-binary-chinese')

# 测试文本(list)
text1 = '这个产品真的太棒了，我非常喜欢！'
text2 = '这次体验太糟糕了，我再也不会购买了。'
texts = [text1,text2]

# 进行情感分析
results = classifier(texts)

# 输出结果
for text,result in zip(texts,results):
    print(f'文本:{text}, 情感:{result["label"]}, 置信度:{result["score"]:.4f}')
```

#### 利用原始的 BERT 预训练模型进行序列分类任务（这里是情感分析）

**logits = outputs.logits** [提取模型的原始输出]

Logits (原始分数)： 在分类任务中，模型最后一层（分类头）通常会为每个可能的类别输出一个原始分数。这些分数还没有经过 Softmax 归一化，因此它们不是概率，可以为任意实数值（正数或负数）。

对于这个代码的二分类任务 (num_labels=2)，logits 的形状通常是 [Batch Size, 2]，即：

第一列代表类别 0（例如：负面）的原始分数。

第二列代表类别 1（例如：正面）的原始分数。

**predictions = torch.argmax(logits, dim = -1)** [确定最终预测类别]
- torch.argmax(): 这是一个 PyTorch 函数，用于找到给定张量中最大值所在的索引。在分类任务中，这个索引就是模型的预测类别 ID。

- logits: 作为输入张量，即我们刚提取的原始分数。

- dim = -1: 这是一个关键参数，指定了我们要查找最大值的维度。

    在 PyTorch 中，dim = -1 指的是最后一个维度。

    在我们的 [Batch Size, 2] 形状中，最后一个维度（大小为 2）代表类别。

    通过沿着类别维度查找最大值，我们实际上是在问：“在所有类别的分数中，哪个分数最高？”

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# 1️. 加载预训练的 BERT 中文分词器 & 模型:这里的 BERT 只是加载了未经微调的权重。虽然模型能运行，但它还没有学习如何区分“正面”和“负面”情感。它的预测结果将是随机的，因为分类头尚未经过训练
# 加载一个预训练的 BERT 分词器，它可以将输入文本转换为 BERT 模型所需的 token ID。
tokenzier = BertTokenizer.from_pretrained('uer/roberta-base-finetuned-jd-binary-chinese')
# 加载一个 BERT 预训练模型，并将其用于二分类任务（如情感分析）
model = BertForSequenceClassification.from_pretrained('uer/roberta-base-finetuned-jd-binary-chinese',num_labels = 2) # 适配于二分类（num_labels=2）任务

# 2️. 测试样例文本
texts = ['这个电影太好看了！','这家餐厅的服务太差了。']

# 3️.分词 & 编码
inputs = tokenzier(texts,
                   padding = True,          # 对较短的句子进行填充（Padding），使其与批次中最长的句子长度一致，方便批量处理
                   truncation = True,       # 如果任何句子超过了模型的最大长度限制（通常是 512），则对其进行截断
                   return_tensors = 'pt')   # 指定输出结果为 PyTorch 张量

# 4️. 进行推理
with torch.no_grad():                       # 这是在进行推理（而不是训练）时的标准做法。它会禁用梯度计算，从而节省内存并加速计算
    outputs = model(**inputs)               # 使用 Python 的解包语法将 inputs 字典中的键值对（如 input_ids, attention_mask 等）作为命名参数传入模型
    # outputs 接收模型的输出，它是一个包含 logits（原始预测分数）等信息的对象。

# 5️. 获取预测结果
logits = outputs.logits # .logits: 这是一个属性，用于提取模型的原始输出分数
predictions = torch.argmax(logits,dim = -1)
for text,pred in zip(texts,predictions): # zip(texts, predictions): 这是 Python 的一个内置函数，它将两个或多个列表按元素索引位置进行配对组合，并创建一个可迭代的元组
    sentiment = '正面' if pred.item() == 1 else '负面'
    # if pred.item() == 1: 如果预测的索引是 $1$，则将 sentiment 变量赋值为 '正面'。
    # else '负面': 如果预测的索引不是 $1$（即是 $0$，因为设置了 num_labels=2），则将 sentiment 变量赋值为 '负面'
    print(f"文本: {text} -> 预测情感: {sentiment}")
```

### 实战 - 批量情感分析打分

#### 读取文件excel

```python
import pandas as pd

data = pd.read_excel('weibo_data1.xlsx')
data.head()
```

将评论转化为列表lists

```python
comments = [data.iloc[i,3] for i in range(len(data))]
comments
```

#### 情感分析

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# 加载预训练模型
tokenzier = BertTokenizer.from_pretrained('uer/roberta-base-finetuned-jd-binary-chinese')
model = BertForSequenceClassification.from_pretrained('uer/roberta-base-finetuned-jd-binary-chinese') 

# 设定模型为评估模式
model.eval()
# 将模型设置为评估模式 (eval())。这会关闭训练时特有的功能（如 Dropout、Batch Normalization 层的更新），确保模型在推理时行为稳定且输出确定

# 处理每条评论并进行推理
scores= []
for comment in comments:
    inputs = tokenizer(comment,
                       padding = True,
                       truncation = True,
                       max_length = 512,
                       return_tensors = 'pt')
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        # 从 outputs 中提取出 logits，这是模型对于每个类别的原始得分。在二分类模型中，logits 是一个包含两个数值的向量，对应正面和负面情绪的原始得分。
        
        # softmax 函数将 logits 转换为概率值，使得输出结果的所有类别的概率之和为 1。
        prob = torch.nn.functional.softmax(logits,dim = -1) # 计算正/负面概率
        scores.append(prob[0][1].item()) # 取正面情绪的概率
        # prob[0]: 由于输入是一个批次（Batch）大小为 1 的张量，prob[0] 取出了这一个样本的概率向量。
        # [1]: 提取向量中索引为 1 的概率值。根据模型约定，这个索引通常代表正面情绪的概率 $P(\text{正面})$。
        # .item(): 将 PyTorch 张量中的数值提取为标准 Python 浮点数，并存入 scores 列表。

# 统计正面和负面评论数量
positive = []
negative = [] 
for score in scores:
    if (score > 0.5):
        positive.append(score)
    else:
        negative.append(score)

print(scores)

# 计算整体情感得分（0-1 之间，越高越正面）
average_score = sum(scores)/len(scores)
print(f"整体情感得分: {average_score:.2f}")
```
