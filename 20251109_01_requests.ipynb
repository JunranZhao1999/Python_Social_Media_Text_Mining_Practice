```python
html = """
<html>
    <head><title>Test Page</title></head>
    <body>
        <h1>This is a test header</h1>
        <p>This is a paragraph.</p>
    </body>
<html>
"""

# 解析html元素
# 使用html.parser解析器
soup = BS(html,'html.parser')
print(soup)

# 查找并打印标题
print(soup.title.text)
print(soup.h1.text)
print(soup.p.text)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# 启动浏览器
driver = webdriver.Chrome()

# 最大化窗口
driver.maximize_window()

# 访问网页
driver.get('https://www.baidu.com')

# 输入关键词 # By 是 Selenium 中用来定位页面元素的方法。常用的定位方式包括 ID、名称、类名、标签名、CSS 选择器、XPath 等
input_element = driver.find_element(By.CLASS_NAME,'s_ipt') # 根据类名's_ipt'定位输入框
input_element.send_keys("北京公园") # send_keys 自动填充

# 确定搜索 
search_btn = driver.find_element(By.CSS_SELECTOR,'.bg.s_btn') # 找到search button
# bg s_btn 这两个类名是 两个独立的类，By.CLASS_NAME 只支持单个类。
search_btn.click() # 点击确定

# 关掉浏览器
# driver.quit()
```

```python
# 获取小红书的cookie

def GetCookies(url):

    driver = webdriver.Chrome()

    # 打开网页
    driver.get(url)

    # 留出输入密码的时间
    time.sleep(50)

    # 获取更新的cookies
    driver.refresh()
    cookies = driver.get_cookies()

    # 输出cookies
    for cookie in cookies:
        print(cookie)

    # 将cookies转为header格式 : 'Apache':'349490162003.9359.1738739237863'
    cookie_string = ':'.join([f'{cookie['name']} = {cookie['value']}' for cookie in cookies])
    print(cookie_string)

url = 'https://xiaohongshu.com'
GetCookies(url)
```

```python
# 数据结构
# `cookies = [{'name': 'session_id', 'value': 'abc'}, {'name': 'lang', 'value': 'en'}]`

# f'{cookie["name"]} = {cookie["value"]}'
# ['session_id = abc', 'lang = en']

# 外部的 .join() 方法用于将列表推导式生成的字符串列表连接成一个单一的字符串。
# 将列表中的所有字符串元素连接起来，并在每个元素之间插入 :
# 'session_id = abc:lang = en'
```
